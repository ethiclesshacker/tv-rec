```python
import numpy as np
import pandas as pd
import itertools
```

```python
data = pd.read_table("data\\imdb-title-cast-rating-desc.tsv")
data = data.dropna()
data = data[data["genre"]!="\\N"]
```

```python
show_genres = list(data["genre"])

genres = ",".join(show_genres)
genres = genres.split(",")
genres = set(genres)
```

```python

print(len(genres))
print(genres)
```

```python
sum = 0
sqs = 0
for genre in genres:
    df = data[data["genre"].str.contains(genre)]
    rows = df.shape[0]
    # print(genre, rows)
    sum += rows 
    sqs += rows * rows
print("Sum =",sum)
print("Sq =",sqs)
```

```python
mega_set = set()


for genre in genres:
    print(genre)
    df = data[data["genre"].str.contains(genre)]
    titles = df["title"]
    for t1 in titles:
        for t2 in titles:
            if t1 != t2:
                mega_set.add(tuple(sorted([t1,t2])))

```
