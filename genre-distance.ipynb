```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_table("tvs-data.tsv")
data = data[data["genres"]!="\\N"]
data.reset_index()
```

```python
show_genres = list(data["genres"])

genres = ",".join(show_genres)
genres = genres.split(",")
genres = set(genres)
```

```python

print(len(genres))
print(genres)
```

```python
distance = list()
for _ in genres:
    distance.append([])

for i,genre in enumerate(genres):
    for sec_genre in genres:
        if [genre == sec_genre]:
            distance[i].append(0)
            continue
        df = data[data["genres"].str.contains(genre) & data["genres"].str.contains(sec_genre)]
        distance[i].append(df.shape[0])
        # print(i,genre,j,sec_genre)
        # print(df.shape)
        
```

```python
dist = pd.DataFrame(distance,index=genres,columns=genres)
```

```python
dists = {}
for genre in genres:
    dists[genre] = {}

for genre in genres:
    for sec_genre in genres:
        
        if sec_genre in dists[genre].keys():
            continue

        if genre == sec_genre:
            dists[genre][sec_genre] = 0
            dists[sec_genre][genre] = 0
            continue

        df = data[data["genres"].str.contains(genre) & data["genres"].str.contains(sec_genre)]
        dists[genre][sec_genre] = df.shape[0]
        dists[sec_genre][genre] = df.shape[0]
```

```python
distance_df = pd.DataFrame.from_dict(dists)
distance_df
```
