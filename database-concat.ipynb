```python
import numpy as np
import pandas as pd
```

```python
main_data = pd.read_table("tvs-data.tsv")
main_data = main_data[main_data["genres"]!="\\N"]
# main_data
```

```python
principal_data = pd.read_table("principal-data.tsv",usecols=["tconst","nconst"])
# principal_data
```

```python
tvlists = list(main_data["tconst"])
# print(tvlists)
```

```python
prc = principal_data[principal_data["tconst"].isin(tvlists)]
# prc
```

```python
prc['cast'] = prc.groupby(['tconst'])['nconst'].transform(lambda x : ' '.join(x))
```

```python
prc = prc.drop(["nconst"],axis=1)
prc = prc.drop_duplicates()
```

```python
prc
```

```python
mdata = pd.merge(main_data,prc,on="tconst")
mdata.drop_duplicates()
mdata
```

```python
mdata.to_csv("data-with-cast.tsv",sep="\t")
```
